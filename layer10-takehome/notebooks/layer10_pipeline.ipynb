# 🧠 Layer10 — Grounded Long-Term Memory Pipeline
### Structured Extraction · Deduplication · Memory Graph · Retrieval · Visualization
**Corpus:** Enron Email Dataset (CMU, 517K emails across 150 inboxes)  
**Model:** Groq Llama-3.3-70B (free tier) + sentence-transformers all-MiniLM-L6-v2  
**Store:** PostgreSQL + pgvector (adjacency model)

---
Reading the problem statement, here's what we actually do with 517K emails:

| Stage | What we do | Why |
|---|---|---|
| **Corpus** | Walk maildir, filter to 5 key inboxes (Lay, Skilling, Fastow…) | 517K→~15K manageable emails |
| **Dedup** | SHA-256 body hash + MinHash near-dup → canonical clusters | Enron has massive reply-quoting |
| **Extract** | LLM → typed entities + claims + verbatim excerpts per email | Ground every fact to a source |
| **Validate** | Pydantic schema + repair + confidence gate (≥0.4) | No noise enters durable memory |
| **Canonicalize** | Embed entities → cosine dedup → merge aliases | "Jeff Skilling" = "J. Skilling" |
| **Graph** | Entities ↔ Claims (edges) ↔ Evidence (pointers) in Postgres | Queryable, time-aware, grounded |
| **Retrieve** | Embed question → ANN → expand graph → context pack | Every answer traces to evidence |
| **Visualize** | sigma.js Web UI + PyVis notebook view | Explorable, filterable |
